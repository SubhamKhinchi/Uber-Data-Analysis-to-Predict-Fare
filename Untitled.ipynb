# Data Manipulation

### Let's convert our Time_stamp column into date time formate

```python
cab_data['datetime'] = pd.to_datetime(cab_data['time_stamp'], errors='coerce')
weather_data['date_time'] = pd.to_datetime(weather_data['time_stamp'], errors='coerce')
```
